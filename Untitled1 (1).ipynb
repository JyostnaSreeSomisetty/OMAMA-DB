```python
import json
import pandas as pd
import tarfile
import os
from collections import defaultdict
from google.colab import drive
```

```python
# Mount Google Drive and define path
drive.mount('/content/drive')

tar_file_path = '/content/drive/MyDrive/jyo/256x256.tar'  # Update this to match your tar file location
```

```python
# Extract the tar file
print("Extracting the tar file...")
extracted_dir = './extracted'
os.makedirs(extracted_dir, exist_ok=True)

with tarfile.open(tar_file_path, 'r') as tar:
    tar.extractall(extracted_dir)
print(f"Extraction complete! Files are extracted to {extracted_dir}")
```

```python
# Rename the extracted files for consistency
print("Renaming extracted files...")
for i, filename in enumerate(os.listdir(extracted_dir)):
    # Get the file extension
    extension = os.path.splitext(filename)[1]


    new_filename = f"file_{i+1}{extension}"
    os.rename(os.path.join(extracted_dir, filename), os.path.join(extracted_dir, new_filename))
print("Renaming complete!")
```

```python
# Locate the JSON files recursively in the extracted folder
json_files = []
for root, _, files in os.walk(extracted_dir):
    for file in files:
        if file.endswith('.json'):
            json_files.append(os.path.join(root, file))

if len(json_files) == 0:
    raise FileNotFoundError("No JSON files found in the extracted directory or its subdirectories!")

print(f"Found {len(json_files)} JSON file(s). Processing the first file: {json_files[0]}")
```

```python
# Process JSON files to count views and views per patient

view_counts = defaultdict(int)
patient_views = defaultdict(set)

def process_json_file(file_path):
    global view_counts, patient_views

    with open(file_path, 'r') as file:
        for line in file:  # Assuming each line is a valid JSON object
            data = json.loads(line.strip())

            # Extract patient ID, view, and laterality
            patient_id = data.get("PatientID")
            view = data.get("View")
            laterality = data.get("ImageLaterality")

            # Skip invalid entries
            if not all([patient_id, view, laterality]):
                continue

            # Standardize the view and laterality
            view = view.strip().lower().replace("cranio-caudal", "craniocaudal")
            laterality = laterality.strip().upper().replace("LEFT", "L").replace("RIGHT", "R")
            full_view = f"{view.capitalize()} - {laterality}"

            # Update counts
            view_counts[full_view] += 1
            patient_views[patient_id].add(full_view)

# Process all JSON files
print("Processing JSON files...")
for json_file in json_files:
    process_json_file(json_file)
print("Processing complete!")
```

```python
# Summarize the results
print(f"Total distinct views: {len(view_counts)}")
print("View breakdown:")
for view, count in view_counts.items():
    print(f"  {view}: {count}")

# Count views per patient
patient_views_summary = {patient: len(views) for patient, views in patient_views.items()}
print(f"\nTotal patients: {len(patient_views_summary)}")
print(f"Example patient views count: {list(patient_views_summary.items())[:5]}")  # Show first 5 patients
```

```python
# Save results
view_counts_df = pd.DataFrame(view_counts.items(), columns=["View", "Count"])
patient_views_df = pd.DataFrame(patient_views_summary.items(), columns=["PatientID", "Number of Views"])

view_counts_df.to_csv("view_counts.csv", index=False)
patient_views_df.to_csv("patient_views.csv", index=False)

print("Results saved as 'view_counts.csv' and 'patient_views.csv'.")
```
